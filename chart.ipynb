```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def plotSerial(sort):
    data = pd.read_csv(f"./{sort}_Serial.csv")

    plt.plot(data['Tamanho'], data['Tempo'])
    plt.xlabel('Tamanho')
    plt.ylabel('Tempo ()')
    plt.title(f'{sort} Serial')

    plt.show()

def plotParallel(sort):
    data = pd.read_csv(f"./{sort}_Parallel.csv", sep=',')
    data['Tempo'] = data.groupby(['Tamanho', 'Threads'])['Tempo'].transform('mean')

    for threads in data['Threads'].unique():
        plt.plot(data[data['Threads'] == threads]['Tamanho'], data[data['Threads'] == threads]['Tempo'], label=f'{threads} Threads')
    
    plt.xlabel('Tamanho')
    plt.ylabel('Tempo')
    plt.title(f'{sort} Paralelo')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plotSerial("BubbleSort")
plotParallel("BubbleSort")
```

```python
plotSerial("CountingSort")
plotParallel("CountingSort")
```

```python
plotSerial("MergeSort")
plotParallel("MergeSort")
```

```python
plotSerial("QuickSort")
plotParallel("QuickSort")
```
